# Day-1 for practicing Numpy
  Start with basic

```python
import numpy as np
```

### Making array from list

```python
arr_1=np.array([1,2,3,4,5,6])
print("Array-1:",arr_1)
arr_2=np.array([[1,2,3],[4,5,6]])
print("Array-2:",arr_2)
```

### List VS numpy Array

```python
List=[1,2,3,4,5,6]
print("Multiplication of list:",List*2)
np_array=np.array([1,2,3,4,5,6])
print("Multiplication of numpy array:",np_array*2)
```

### Zeros , ones , random , full , sequence

```python
Zeros=np.zeros((2,2))
print("Zero matrix: \n",Zeros)
Ones=np.ones((5,5))
print("One matrix: \n",Ones)
Full=np.full((4,4),7)
print("Full matrix: \n",Full)
Random=np.random.random((3,3))
print("Random Matrix:\n",Random)
Sequence=np.arange(0,11,2)
print("Sequence:\n",Sequence)
```

### Vector,Matrix and Tensor

```python
Vector=np.array([1,2,3,4,5,6])
print("Vector:\n",Vector)
Matrix=np.array([[1,2,3],[4,5,6]])
print("Matrix:\n",Matrix)
Tensor=np.array([[[1,2],[3,4]],
                 [[5,6],[7,8]]])
print("Tensor:\n",Tensor)
```

### Array Property

```python
arr=np.array([[1,2,3],
             [4,5,6]])
print("Array Shape:",arr.shape)
print("Array Dimension:",arr.ndim)
print("Array Size:",arr.size)
print("Array Datatype:",arr.dtype)
print("Array Base",arr.base)
print("Array Data",arr.data)
```

### Manipulation of size of array

```python
ex=np.arange(0,12)
print("original Matrix:\n",ex)
Reshaped=ex.reshape((2,6))
print("Reshaped Matrix:\n",Reshaped)
Falltened=Reshaped.flatten()
print("Flattened Matrix:\n",Falltened)
Ravel=Reshaped.ravel()
print("Ravel Matrix:\n",Ravel)
Transpose=Reshaped.T
print("Transpose Matrix:\n",Transpose)
```
